```python
%%time
df = dd.read_csv('../data/itineraries.csv', blocksize='5MB', dtype={'totalTravelDistance': 'float64'})
print(df.shape)
df.head()
```

```python
%%time
df.columns
```

Lest's see how much time takes to perform some basic tasks in a large dask dataframe.

These tasks can be done faster with a GPU cluster as you could see in this [video](https://www.youtube.com/watch?v=auBaWD31FCk) 

```python
%%time
df.shape[0].compute()
```

```python
%%time
for col in df.columns:
    print(col, ':', df[col].dtype)
```

```python
%%time
df.count().compute()
```

```python
%%time
df.groupby('startingAirport').agg({'flightDate': 'count', 'totalTravelDistance': np.mean}).compute()
```

As you can see single tasks take more than 5 minutes each one. Then if you are going to take huge datasets make sure to have a GPU with enough memory.
